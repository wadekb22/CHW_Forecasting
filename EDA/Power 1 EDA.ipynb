```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Power Data 1.csv", index_col='Date', parse_dates=True)
data.head()
```

```python
data.isna().sum()
```

```python
data['Total Power'].plot(color='deepskyblue')
plt.xlabel('Month')
plt.ylabel('Total Power')
plt.title('Total Power: March 2019-20')
plt.show()
```

```python
data[data["Total Power"] < 0].count()
```

```python
data[data["Total Power"] < 0]
```

```python
negative_values = data[data["Total Power"] < 0]["Total Power"]

negative_values.plot(style='r.')

plt.show()
```

```python
no_neg_data = data["2019-08"][["Total Power"]]

no_neg_data =  no_neg_data.asfreq("h")

no_neg_data
```

```python
rolling_data = no_neg_data.rolling(12, center=True)

fig = plt.figure()
plt.plot(no_neg_data,'r-',label="Total Power")
plt.plot(rolling_data.mean(),'b-',label="Rolling mean")
plt.plot(rolling_data.std(),'g-',label="Rolling std")
plt.title("Total Power for August 2019")

fig.autofmt_xdate()
plt.legend()
```
