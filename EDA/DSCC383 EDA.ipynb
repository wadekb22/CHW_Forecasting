```python
import pandas as pd

ts = pd.read_csv("merged_data.csv", index_col='DATE', parse_dates=True)
ts.head()
```

```python
ts.drop('Unnamed: 0', axis=1, inplace=True)
ts.head()
```

```python
ts.columns
```

```python
from dateutil.parser import parse

ts2 = pd.read_csv("merged_data.csv")

date_series = ts2["DATE"]
print("Original Series:")
print(date_series)
date_series = date_series.map(lambda x: parse(x))
print("Day of week:")
weekday = list(date_series.dt.day_name())

ts["Weekday"] = weekday
```

```python
ts["Weekday"] = weekday
ts.head()
```

```python
daily_max = ts[["Total Power (max)", "Weekday"]].resample('D').max()
daily_max.tail()
```

```python
spring_dm = daily_max["2020-03-01":"2020-06-01"]
summer_dm = daily_max["2020-06-01":"2020-09-01"]
fall_dm = daily_max["2020-09-01":"2020-12-01"]
winter_dm = daily_max["2020-12-01":"2021-02-28"]
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
#sns.set_theme(style="whitegrid")
#tips = sns.load_dataset("tips")
ax = sns.barplot(x="Weekday", y="Total Power (max)", data=spring_dm)
```

```python
plt.figure(figsize=(10,5))
chart = sns.barplot(x="Weekday", y="Total Power (max)", data=spring_dm)

chart.set_xticklabels(
    chart.get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'
    
)

chart.set_title("Average Max Power Distributed by Week in the Spring",fontsize=20)
chart.set_ylabel("Max Power (tons)",fontsize=18)
chart.set_xlabel("Weekday (a)",fontsize=18)
```
